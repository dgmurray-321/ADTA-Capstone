```python
def value_count_to_barplot(data, column, size, normalize):
    if normalize:
        data_value_count = data[column].value_counts(normalize = True).reset_index(name = 'percentage').sort_values(by = 'percentage')
        plt.figure(figsize=(size))
        ax = sns.barplot(data = data_value_count, x = 'percentage', y = 'index')
        for container in ax.containers:
            ax.bar_label(container)
        plt.title(f'count of {column} (normalized)')
        plt.show()
    else: 
        data_value_count = data[column].value_counts().reset_index(name = 'count').sort_values(by = 'count')
        plt.figure(figsize=(size))
        ax = sns.barplot(data = data_value_count, x = 'count', y = 'index')
        for container in ax.containers:
            ax.bar_label(container)
        plt.title(f"count of {column}")
        plt.show()
```

```python
def make_barplot(data,x, y, size):
    plt.figure(figsize=(size))
    ax = sns.barplot(data = data, x = x, y = y)
    for container in ax.containers:
        ax.bar_label(container)
    plt.show()
```

```python
def make_lineplot(data, time_stamp_column,y, size, **kwargs):
    plt.figure(figsize=size)
    sns.lineplot(data = data, x = time_stamp_column, y = y, **kwargs)
    plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = abs(dataset.corr())
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in dataset.columns:
                    del dataset[colname] # deleting the column from the dataset

    return list(col_corr)
```

```python
columns_removed_based_on_multicollinearity = correlation(dataset=df, threshold=.70)
print("here are the suggested columns to be removed based on correlation/multicollinearity:",  columns_removed_based_on_multicollinearity, '\n')
print("number of columns removed based on correlation/multicollinearity:", len(columns_removed_based_on_multicollinearity))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
ts_ng = df.copy()
ts_ng.set_index('Local time', inplace = True)
seasonal_result = seasonal_decompose(x = ts_ng['NG'], model = 'additive', period = 24*365)
seasonal_result.plot()

```

```python

sns.lineplot(seasonal_result.resid)
```

```python
seasonal_result.resid.describe()
```

```python
sns.histplot(seasonal_result.resid)
```

```python
import scipy.stats as stats
plt.figure(figsize=(14,8))
stats.probplot(seasonal_result.resid, dist = "norm", plot = plt)
plt.show()
```

```python
residuals = seasonal_result.resid.dropna()
shapiro_test = stats.shapiro(residuals)
print("P value for shapiro test:", shapiro_test.pvalue)
km_test = stats.kstest(residuals, "norm", args= (residuals.mean(), residuals.std()))
print("P value for KM test:", km_test.pvalue)
```

The residuals do deviate from the mean, indicating that the decomposition does effectively capture the underlying patterns in the data
